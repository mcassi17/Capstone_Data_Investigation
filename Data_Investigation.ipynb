```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime 
%matplotlib inline
import seaborn as sb
```

```python
bikeshare = pd.read_csv('bikeshare_reduced.csv', 
                        parse_dates=['Start date', 'End date', 'start_date_short', 'end_date_short'])
```

```python
bikeshare.shape
```

```python
bikeshare.head()
```

```python
bikeshare.tail()
```

```python
bikeshare.drop('Unnamed: 0', 1, inplace=True)
```

```python
bikeshare_date_count = bikeshare.groupby('start_date_short')['mnth'].count()
bikeshare_date_count.head(20)
```

```python
bikeshare_date_count.tail(20)
```

```python
bikeshare_date_count.plot()
plt.show()
```

```python
bikeshare_points = bikeshare.copy()
bikeshare_points = bikeshare_points[['Start station', 'End station']]
bikeshare_points['Combo'] = bikeshare_points['Start station'] + ' + ' + bikeshare_points['End station']
bikeshare_points.head()
```

```python
bikeshare_combo = bikeshare_points.groupby('Combo')['Combo'].count().sort_values(ascending=False)
bikeshare_combo.head(20)
```
